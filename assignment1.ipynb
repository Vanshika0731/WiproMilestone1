```python
# Write a program to print the sum of all the digits of a given number.
# Example: I/P: 1234 → O/P: 10

num = 1234
digit_sum = sum(int(d) for d in str(num))
print(f"Sum of digits of {num}: {digit_sum}")


# Write a program to reverse a given number and print.
# Example 1: I/P: 1234 → O/P: 4321
# Example 2: I/P: 1004 → O/P: 4001

num1 = 1234
num2 = 1004
reversed1 = int(str(num1)[::-1])
reversed2 = int(str(num2)[::-1])
print(f"Reverse of {num1}: {reversed1}")
print(f"Reverse of {num2}: {reversed2}")


# Write a program to find if the given number is palindrome or not.
# Example 1: I/P: 110011 → O/P: 110011 is a palindrome.
# Example 2: I/P: 1234  → O/P: 1234 is not a palindrome.

for num in [110011, 1234]:
    s = str(num)
    if s == s[::-1]:
        print(f"{num} is a palindrome.")
    else:
        print(f"{num} is not a palindrome.")
```

```python
# Write a function to return the sum of all numbers in a list.

def sum_list(lst):
    total = 0
    for n in lst:
        total += n
    return total

print(sum_list([8, 2, 3, 0, 7]))


# Write a function to return the reverse of a string.

def reverse_string(s):
    return s[::-1]

print(reverse_string("1234abcd"))


# Write a function to calculate and return the factorial of a number.

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))


# Write a function that accepts a string and prints the number of upper and lower case letters.

def count_cases(s):
    upper = 0
    lower = 0
    for ch in s:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1
    print("Upper:", upper)
    print("Lower:", lower)

count_cases("Hello World")


# Write a function to print the even numbers from a given list.

def print_evens(lst):
    evens = []
    for n in lst:
        if n % 2 == 0:
            evens.append(n)
    print(evens)

print_evens([1, 2, 3, 4, 5, 6, 7, 8, 9])


# Write a function that takes a number and checks whether the number is prime or not.

def is_prime(n):
    if n < 2:
        print(n, "is not prime")
        return
    for i in range(2, n):
        if n % i == 0:
            print(n, "is not prime")
            return
    print(n, "is prime")

is_prime(7)
is_prime(10)
```

```python
# Write a program to accept 10 numbers through command line arguments
# and calculate the sum of prime numbers among them.

import sys

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

numbers = sys.argv[1:]
total = 0
for n in numbers:
    if is_prime(int(n)):
        total += int(n)

print("Sum of prime numbers:", total)
```

```python
# Write a program to accept two numbers from the user and perform division.
# If any exception occurs, print an error message or else print the result.

try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    result = a / b
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except ValueError:
    print("Error: Invalid input. Please enter numbers only.")


# Write a program to accept a number and check whether it's prime or not.
# If user enters anything other than number, handle the exception.

try:
    n = int(input("Enter a number: "))
    if n < 2:
        print(n, "is not prime")
    else:
        is_prime = True
        for i in range(2, n):
            if n % i == 0:
                is_prime = False
                break
        if is_prime:
            print(n, "is prime")
        else:
            print(n, "is not prime")
except ValueError:
    print("Error: Please enter a valid number.")


# Write a program to accept the file name to be opened from the user.
# If file exists print contents in title case, else handle the exception.

try:
    filename = input("Enter file name: ")
    f = open(filename, "r")
    content = f.read()
    f.close()
    print(content.title())
except FileNotFoundError:
    print("Error: File not found.")


# Declare a list with 10 integers and ask the user to enter an index.
# Check whether the number at that index is positive or negative.
# If invalid index is entered, handle the exception.

try:
    lst = [10, -5, 3, -8, 7, -1, 4, -9, 2, -6]
    print("List:", lst)
    index = int(input("Enter an index (0-9): "))
    num = lst[index]
    if num > 0:
        print(num, "is a positive number")
    elif num < 0:
        print(num, "is a negative number")
    else:
        print(num, "is zero")
except IndexError:
    print("Error: Index out of range. Please enter index between 0 and 9.")
except ValueError:
    print("Error: Please enter a valid integer index.")
```
