```python
# Server Cost Calculator

cost_per_hour = float(input("Enter the cost per hour ($): "))
budget = float(input("Enter your budget ($): "))

cost_per_day = cost_per_hour * 24
cost_per_week = cost_per_day * 7
cost_per_month = cost_per_day * 30
days = budget / cost_per_day

print("\nCost to operate one server per day: $", round(cost_per_day, 2))
print("Cost to operate one server per week: $", round(cost_per_week, 2))
print("Cost to operate one server per month: $", round(cost_per_month, 2))
print("Number of days one server can operate with $", budget, ":", round(days, 2))
```

```python
#Find the percentange
# Initialize the record dictionary
student_marks = {
    "Krishna": [67.0, 68.0, 69.0],
    "Arjun": [70.0, 98.0, 63.0],
    "Malika": [52.0, 56.0, 60.0]
}

# Get user input
query_name = input("Enter a name: ")

# Check if the name exists in the records
if query_name in student_marks:
    marks = student_marks[query_name]

    # Calculate average: (sum of marks) / (no of subjects)
    average = sum(marks) / len(marks)

    # Output the result formatted to 2 decimal places (optional) or as a whole number
    print(f"Average percentage mark: {average:.2f}")
else:
    print("Student name not found.")
```

```python
#Sort the color
def sort_colors(color_string):
    # Split the string into a list of colors
    colors = color_string.split('-')

    # Sort the list alphabetically
    colors.sort()

    # Join the sorted list back into a hyphen-separated string
    return "-".join(colors)

# Test Cases
print(sort_colors("green-red-yellow-black-white")) # Output: black-green-red-white-yellow
print(sort_colors("PINK-BLUE-TAN-PURPLE"))        # Output: BLUE-PINK-PURPLE-TAN
```

```python
#Playing with names
def ispalindrome(name):
    # Check if the name is the same forwards and backwards
    if name == name[::-1]:
        return "Yes it is a palindrome."
    else:
        return "No it is not a palindrome."

def count_the_vowels(name):
    vowels = "aeiouAEIOU"
    count = sum(1 for char in name if char in vowels)
    return f"No of vowels: {count}"

def frequency_of_letters(name):
    freq = {}
    # Count occurrences of each character (ignoring spaces for cleaner output)
    for char in name:
        if char != " ":
            freq[char] = freq.get(char, 0) + 1

    # Format the dictionary into the required string format
    freq_str = ", ".join([f"{k}-{v}" for k, v in freq.items()])
    return f"Frequency of letters: {freq_str}"
```

```python
#Happiness
import sys

def calculate_happiness():
    # Reading from command line arguments
    if len(sys.argv) < 4:
        return

    # Extracting the three main strings
    str1 = sys.argv[1].split('-')
    str2 = sys.argv[2].split('-')
    str3 = sys.argv[3].split('-')

    # Convert liked and disliked strings to sets for fast lookup
    liked = set(str1)
    disliked = set(str2)

    happiness = 0

    # Process the numbers given to you (String 3)
    for num in str3:
        if num in liked:
            happiness += 1
        elif num in disliked:
            happiness -= 1

    print(happiness)

if __name__ == "__main__":
    calculate_happiness()
```

```python
#Find the select
from collections import Counter
import re

def find_secret_message(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    # 1. Determine Meeting Time
    num_lines = len(lines)
    if num_lines > 12:
        meeting_time = f"{num_lines - 12} PM"
    else:
        meeting_time = f"{num_lines} AM"

    # 2. Determine Meeting Place
    # Combine all lines and find all words (ignoring punctuation/case)
    text = " ".join(lines).lower()
    words = re.findall(r'\b\w+\b', text)

    if words:
        most_common_word, _ = Counter(words).most_common(1)[0]
        meeting_place = f"{most_common_word.capitalize()} Street"
    else:
        meeting_place = "Unknown Street"

    # Output Results
    print(f"Meeting time: {meeting_time}")
    print(f"Meeting place: {meeting_place}")

# Usage
# find_secret_message('sample.txt')
```

```python
#Purchase
import os

def process_purchase():
    file_name = input("Enter the file name: ")

    # Ensure the file name has the correct extension if not provided
    if not file_name.endswith('.txt'):
        file_name += '.txt'

    try:
        with open(file_name, 'r') as file:
            lines = [line.strip() for line in file if line.strip()]

            if not lines:
                print("The file is empty.")
                return

            items_purchased = []
            discount = 0.0

            # Process lines
            for line in lines:
                parts = line.split()
                # Check if it's the discount line (usually the last one)
                if parts[0].lower() == "discount":
                    discount = float(parts[1])
                else:
                    # Item name and price
                    price = float(parts[-1])
                    items_purchased.append(price)

            # Calculations
            no_of_items = len(items_purchased)
            free_items = items_purchased.count(0)
            total_amount = sum(items_purchased)
            final_amount = total_amount - discount

            # Output results
            print(f"No of items purchased: {no_of_items}")
            print(f"No of free items: {free_items}")
            print(f"Amount to pay: {total_amount}")
            print(f"Discount given: {discount}")
            print(f"Final amount paid: {final_amount}")

    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except ValueError:
        print("Error: The file contains invalid data formats (ensure prices are numbers).")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    process_purchase()
```
